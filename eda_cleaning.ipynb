```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
filtered = pd.read_csv("Street_Tree_List-2022-01-30_FILTERED.csv")
```

```python
filtered
```

```python
# drop NaN for PlantDate
filtered = filtered.dropna(subset=['PlantDate'])
filtered
```

```python
filtered['PlantDate'] = pd.to_datetime(filtered.PlantDate)
filtered['PlantDate_year'] = filtered.PlantDate.dt.year
filtered['PlantDate_month'] = filtered.PlantDate.dt.month
```

```python
# sanity check: There are some years that are >2022
print(sorted(filtered.PlantDate_year.unique()))
```

```python
# sanity check: has all months 
sorted(filtered.PlantDate_month.unique())
```

```python
cleaner = filtered[filtered.PlantDate_year<=2022]
```

```python
print(sorted(cleaner.PlantDate_year.unique()))
```

```python
cleaner
```

```python
cleaner=cleaner.reset_index(drop=True)
```

```python
x="2123"
x[:3]
```

```python
cleaner['PlantDate_decade'] = cleaner['PlantDate_year'].apply(str).str[:3]
```

```python
mapRange = {'197':'1973-1979', '198':'1980-1989', '199':'1990-1999', '200':'2000-2009', '201': '2010-2019', '202': '2020-2022'}
cleaner['PlantDate_decade'] = cleaner['PlantDate_decade'].map(mapRange)
```

```python
decades = cleaner.groupby(['PlantDate_decade', 'PlantDate_month']).count()
```

```python
decades = decades.reset_index()
```

```python
decades.head()
```

```python
sns.lineplot(decades, x='PlantDate_month', y='TreeID', hue='PlantDate_decade')
```

```python
decades.columns = ['TreeCount' if x=='count' else x for x in decades.columns]
```

```python
decadesAvg = decades.iloc[:,:3]
```

```python
decadesAvg
```

```python
concat1=decadesAvg[decadesAvg['PlantDate_decade']=='1973-1979']['TreeCount'].div(7.0)
```

```python
concat3=decadesAvg[decadesAvg['PlantDate_decade']=='2020-2022']['TreeCount'].div(3.0)
```

```python
concat2=decadesAvg.loc[np.where((decadesAvg['PlantDate_decade']!='2020-2022') & (decadesAvg['PlantDate_decade']!='1973-1979'))]['TreeCount'].div(10)
```

```python
#1973-1979 , /7
#2020-2022, /3
#else, /10

```

```python
AvgTreePlantYear = list(concat1.values)+list(concat2.values)+list(concat3.values)
```

```python
decadesAvg['AvgTreePlantYear']=AvgTreePlantYear
```

```python
decadesAvg
```

```python
sns.lineplot(decadesAvg, x='PlantDate_month', y='AvgTreePlantYear', hue='PlantDate_decade')
```

```python
sns.lineplot(decadesAvg[decadesAvg['PlantDate_decade']!='2000-2009'], x='PlantDate_month', y='AvgTreePlantYear', hue='PlantDate_decade')
```

```python
allYears = decadesAvg.groupby('PlantDate_month').sum()
```

```python
sns.lineplot(allYears, x='PlantDate_month', y='AvgTreePlantYear')
```

```python
allYears=pd.DataFrame(allYears).reset_index()
allYears['PlantDate_decade']='1973-2023'
```
